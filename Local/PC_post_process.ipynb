```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
from pathlib import Path

"""Define general plotting parameters"""
plt.rcParams['figure.dpi'] = 100
plt.rcParams.update({"text.usetex": False, 
                     "font.family": "serif", 
                     "mathtext.fontset": "cm", 
                     "font.size": 9, 
                     "axes.labelsize": 9, 
                     "legend.fontsize": 8, 
                     "xtick.labelsize": 8, 
                     "ytick.labelsize": 8})
pagewidth = 6.3 * 1.2
w_plot_s = pagewidth * 0.5
h_plot_s = w_plot_s / 1.6
w_plot_mw = pagewidth * 0.75
h_plot_mw = w_plot_mw / 1.6
w_plot_fw = pagewidth * 1.0
h_plot_fw = w_plot_fw / 1.6
markers = ['o', 's', '^', 'X', 'v', 'P', 'd', 'p', '<']
colors_d = ['#1f4e79', '#a83232', '#d4a017']
colors_l = ['#76b0e0', '#f0827f', '#f2d98c']
```

```python

def add_axes_to_png(input_file, output_file=None, label=None):
    # Load image
    img = mpimg.imread(input_file)
    h_image = 30
    w_image = 30
    # Create figure with axes
    fig, ax = plt.subplots(figsize=(w_plot_s, w_plot_s))

    # Show image as background
    ax.imshow(img, extent=[-w_image/2, w_image/2, -h_image/2, h_image/2])  # extent sets coordinates

    ax.set_xlabel(r'$x \, / \, \delta_T$')
    ax.set_ylabel(r'$y \, / \, \delta_T$', labelpad=-2)
    ax.tick_params(axis='both', which='both', labelcolor='black', direction='in')
    ax.minorticks_on()
    ax.set_xticks([i for i in range(int(-w_image/2), int(w_image/2 + 1), 5)])
    ax.set_yticks([i for i in range(int(-h_image/2), int(h_image/2 + 1), 5)])
    ax.set_aspect('equal')
    ax.margins(x=0, y=0)

    if label is not None: ax.text(-13, 11, label, fontsize=12, fontweight='bold', color="white")
    #fig.tight_layout()
    if output_file is not None:
        fig.savefig(output_file, dpi=300, bbox_inches='tight')


    #plt.savefig(output_file, dpi=300, bbox_inches="tight")
    #plt.close(fig) , pad_inches=0
    #print(f"Saved {output_file}")
#, pad_inches=0
username = os.getlogin()
dest_dir = Path(f"C:/Users/{username}/OneDrive - ETH Zurich/Semester 10/Master Thesis/Report/Figures/Pseudocolor_images")
source_dir = Path(f"C:/Users/{username}/OneDrive - ETH Zurich/Semester 10/Master Thesis/Report/Figures/Pseudocolor_images")

u_in_list = ["01.50", "01.75", "02.00", "02.25", "02.50"]

for i, u_in in enumerate(u_in_list):
    source_path = source_dir / f"OH_Pcol_full_{u_in}_viridis.png"
    dest_path = dest_dir / f"pcyoh_annotated_{u_in}.png"
    # labels = ["a)", "b)", "c)", "d)", "e)", "f)"]
    label = f"{r'$\tilde{\dot{m}}_{in}='}{float(u_in)*3:.2f}{r'$'}"
    add_axes_to_png(source_path, output_file=dest_path, label=label)

u_in_list = ["01.25", "02.75"]
for i, u_in in enumerate(u_in_list):
    source_path = source_dir / f"OH_Pcol_full_{u_in}_viridis.png"
    dest_path = dest_dir / f"pcyoh_annotated_{u_in}.png"
    label = ' '
    #label = f"{r'$\tilde{\dot{m}}_{in}='}{float(u_in)*3:.2f}{r'$'}"
    add_axes_to_png(source_path, output_file=dest_path, label=label)
```
