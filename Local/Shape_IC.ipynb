```python
import numpy as np
import matplotlib.pyplot as plt

def get_n_cell(theta, n, r_mean):
    r = np.ones((len(theta)))
    r = r * r_mean
    a = 1 / 15
    b = a / 3
    c = b / 3
    r = r*(1 + np.cos(n * theta) * a + np.cos(2 * n * theta) * b + np.cos(3 * n * theta) * c)
    print(a, b, c)
    return r
```

```python
# Uin 2.75 l ~ 54

theta = np.linspace(0, 2 * np.pi, 1000)
radius = get_n_cell(theta, 3, 6.8)

ig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, radius)

ax.set_title("Polar Shape from Angles and Radii")
plt.show()
```
